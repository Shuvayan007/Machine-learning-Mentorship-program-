```python
final=pd.DataFrame()

for j in range (1,21):
    
    headers={'user-Agent': 'Mozilla/5.0 (windows NT 6.3; Win64; x64) AppleWebkit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page={}'.format(j), headers=headers).text
    
    soup=BeautifulSoup(webpage, 'lxml')
    
    company=soup.find_all('div',class_='company-content')
    
    name=[]
    rating=[]
    reviews=[]
    ctype=[]
    hq=[]
    old=[]
    employees=[]
    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('p',class_='rating').text.strip())
        reviews.append(i.find('a',class_='review-count').text.strip())

        try:
            ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
        except:
            ctype.append(np.nan)
        try:
            hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
        except:
            hq.append(np.nan)
        try:
            old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
        except:
            old.append(np.nan)
        try:
            employees.append(i.find_all('p',class_='infoEntity')[3].text.strip())
        except:
            employees.append(np.nan)
            
    #    Use try and except, if there are missing values in particular column
    #    In case of missing values same indexing won't work

    d={'name':name,
       'rating':rating,
       'reviews':reviews,
       'type':ctype,
       'hq':hq,
       'old':old,
       'employees':employees}
    df=pd.DataFrame(d)
    
    final=final.append(df,ignore_index=True)
    
```
